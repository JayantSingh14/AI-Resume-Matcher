```python
from sentence_transformers import SentenceTransformer
from utils.xai_explain import sentence_importance, token_importance_loo, plot_token_importance

model = SentenceTransformer("all-MiniLM-L6-v2")

job_desc = "Looking for a Data Scientist skilled in Python, SQL, and Machine Learning."
resume = "Data Scientist with 3 years experience in Python, TensorFlow, SQL, and deep learning model deployment."

# Sentence importance
top_sents = sentence_importance(job_desc, resume, model, top_k=2)
print("Sentence-level Explanation:")
for s in top_sents:
    print(s)

# Token importance
tokens = token_importance_loo(job_desc, resume, model, top_n_tokens=10)
plot_token_importance(tokens, top_k=8)
```
